```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh

# Parameters
N = 81
dx = 0.1
k = 1  # Example constant for k, adjust as needed
L = 4

# Initialize the matrix A1
A1 = np.zeros((N, N))

# Define x_n for each index n
x = np.linspace(-L, L, N)

# Fill the matrix A1
for n in range(N):
    A1[n, n] = -2 - k * (x[n]**2) * (dx**2)  # Center diagonal
    if n > 0:
        A1[n, n-1] = 1  # Lower diagonal
    if n < N-1:
        A1[n, n+1] = 1  # Upper diagonal

# Initialize the matrix A2 with zeros
A2 = np.zeros((N, N))

# Set the specified values in A2
A2[0, 0] = 4/3  # Position (1,1) in 1-based indexing
A2[0, 1] = -1/3  # Position (1,2) in 1-based indexing
A2[-1, -2] = -1/3  # Position (-1,-2) in 1-based indexing
A2[-1, -1] = 4/3  # Position (-1,-1) in 1-based indexing

# Combine matrices A1 and A2 to form A3
A3 = A1 + A2

# Compute eigenvalues and eigenvectors
eigvals, eigvec = eigh(-A3)

# Display the eigenfunctions by plotting the eigenvectors
plt.figure(figsize=(10, 6))
for i in range(4):  # Plot the first 4 eigenfunctions
    plt.plot(x, eigvec[:, i], label=f'Eigenfunction {i+1}')
plt.xlabel('x')
plt.ylabel('Eigenfunction')
plt.title('Eigenfunctions of the Matrix A3')
plt.legend()
plt.grid(True)
plt.show()

print(eigvals)
```
